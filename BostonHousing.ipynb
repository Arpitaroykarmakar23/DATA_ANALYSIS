```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn import preprocessing
from sklearn import datasets
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, mean_squared_error,r2_score
```

```python
data=pd.read_csv('/content/BostonHousing.csv')
data
```

```python

X_new= data
y_new= data['medv']

X_new_train, X_new_test, y_train, y_test= train_test_split(X_new ,y_new, test_size= .2, random_state= 42)
print(X_new_train.shape, X_new_test.shape, y_train.shape, y_test.shape)
```

```python
X= data['indus']
Y= data['medv'].values
X_train, X_test, Y_train, Y_test= train_test_split(X, Y, test_size= .2, random_state= 42)
print(X_train.shape, X_test.shape, Y_train.shape, Y_test.shape)
print(Y)
```

```python
model=LinearRegression()
model.fit(X_train.values.reshape(-1,1),Y_train)
```

```python
Y_pred=model.predict(X_test.values.reshape(-1,1))
```

```python
plt.scatter(X_test,Y_test,color='black')
plt.plot(X_test,Y_pred,color='red',linewidth='3')
plt.xlabel("crim")
plt.ylabel("medv")
plt.title("Linear Regression Model")
```

```python
mae= mean_absolute_error(Y_test,Y_pred)
mse=mean_squared_error(Y_test,Y_pred)
rmse=np.sqrt(mse)
r_sq=r2_score(Y_test,Y_pred)
print("Ma", mae)
print("Ms",mse)
print("Rm",rmse)
print("Rsq",r_sq)
```
